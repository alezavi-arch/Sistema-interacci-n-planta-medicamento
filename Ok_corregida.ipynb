```python
# -*- coding: utf-8 -*-
"""
Sistema Interactivo de Prevenci√≥n de Interacciones entre Plantas Medicinales y F√°rmacos
"""

# üß™ BASE DE DATOS CON LAS PLANTAS SOLICITADAS
plantas_db = [
    {
        "Nombre Cient√≠fico": "Croton lechleri",
        "Nombre Com√∫n": "Sangre de Drago",
        "Familia": "Euphorbiaceae",
        "Principios Activos": "Taspina, dimetilcedrusina, proantocianidinas",
        "Usos Tradicionales": "Cicatrizante, antiinflamatorio, antiviral",
        "Mecanismo Interacci√≥n": "Inhibici√≥n del citocromo P450",
        "Interacciones": {
            "Warfarina": "üî¥ ALTO RIESGO: Hemorragias (Evidencia: 24 casos)",
            "Aspirina": "üî¥ ALTO RIESGO: Potenciaci√≥n anticoagulante",
            "Heparina": "üü° RIESGO MODERADO: Aumento tiempo de sangrado"
        }
    },
    {
        "Nombre Cient√≠fico": "Banisteriopsis caapi",
        "Nombre Com√∫n": "Yag√©",
        "Familia": "Malpighiaceae",
        "Principios Activos": "Harmina, harmalina, tetrahidroharmina",
        "Usos Tradicionales": "Medicina tradicional, tratamiento de depresi√≥n",
        "Mecanismo Interacci√≥n": "Inhibici√≥n de la monoaminooxidasa",
        "Interacciones": {
            "Antidepresivos ISRS": "üî¥ ALTO RIESGO: S√≠ndrome serotonin√©rgico",
            "Dextrometorfano": "üî¥ ALTO RIESGO: Toxicidad serotonin√©rgica"
        }
    },
    {
        "Nombre Cient√≠fico": "Ilex guayusa",
        "Nombre Com√∫n": "Guayusa",
        "Familia": "Aquifoliaceae",
        "Principios Activos": "Cafe√≠na, teobromina, L-teanina",
        "Usos Tradicionales": "Estimulante, antioxidante, regulador gluc√©mico",
        "Mecanismo Interacci√≥n": "Inhibici√≥n CYP1A2",
        "Interacciones": {
            "Clozapina": "üî¥ ALTO RIESGO: Toxicidad aumentada",
            "Teofilina": "üü° RIESGO MODERADO: Potenciaci√≥n efectos"
        }
    },
    {
        "Nombre Cient√≠fico": "Gliricidia sepium",
        "Nombre Com√∫n": "Matarrat√≥n",
        "Familia": "Fabaceae",
        "Principios Activos": "Cumarinas, flavonoides, alcaloides",
        "Usos Tradicionales": "Antiparasitario, insecticida, tratamiento de piel",
        "Mecanismo Interacci√≥n": "Inhibici√≥n CYP3A4",
        "Interacciones": {
            "Estatinas": "üî¥ ALTO RIESGO: Rabdomi√≥lisis",
            "Anticoagulantes": "üü° RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Cient√≠fico": "Maytenus laevis",
        "Nombre Com√∫n": "Chuchuguaza",
        "Familia": "Celastraceae",
        "Principios Activos": "Maytansina, triterpenos, polifenoles",
        "Usos Tradicionales": "Afrodis√≠aco, antiinflamatorio, tratamiento reumatismo",
        "Mecanismo Interacci√≥n": "Modulaci√≥n sistema inmunol√≥gico",
        "Interacciones": {
            "Inmunosupresores": "üî¥ ALTO RIESGO: Antagonismo terap√©utico",
            "Corticosteroides": "üü° RIESGO MODERADO: Reducci√≥n eficacia"
        }
    },
    {
        "Nombre Cient√≠fico": "Petiveria alliacea",
        "Nombre Com√∫n": "Anam√∫",
        "Familia": "Phytolaccaceae",
        "Principios Activos": "Flavonoides, cumarinas, triterpenos",
        "Usos Tradicionales": "Antitumoral, analg√©sico, inmunoestimulante",
        "Mecanismo Interacci√≥n": "Inducci√≥n enzim√°tica CYP450",
        "Interacciones": {
            "Anticoagulantes": "üî¥ ALTO RIESGO: Sangrados",
            "Quimioterap√©uticos": "üü° RIESGO MODERADO: Potencial interacci√≥n"
        }
    },
    {
        "Nombre Cient√≠fico": "Cissus verticillata",
        "Nombre Com√∫n": "Insulina Vegetal",
        "Familia": "Vitaceae",
        "Principios Activos": "Fitosteroles, flavonoides, √°cidos grasos",
        "Usos Tradicionales": "Antidiab√©tico, hipoglucemiante, antiinflamatorio",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n insulina",
        "Interacciones": {
            "Insulina": "üî¥ ALTO RIESGO: Hipoglucemia severa",
            "Metformina": "üü° RIESGO MODERADO: Potenciaci√≥n efecto"
        }
    },
    {
        "Nombre Cient√≠fico": "Uncaria tomentosa",
        "Nombre Com√∫n": "U√±a de Gato",
        "Familia": "Rubiaceae",
        "Principios Activos": "Alcaloides oxind√≥licos, glic√≥sidos, taninos",
        "Usos Tradicionales": "Inmunomodulador, antiinflamatorio, antiviral",
        "Mecanismo Interacci√≥n": "Estimulaci√≥n sistema inmunol√≥gico",
        "Interacciones": {
            "Inmunosupresores": "üî¥ ALTO RIESGO: Antagonismo terap√©utico",
            "Vacunas vivas": "üü° RIESGO MODERADO: Respuesta inmunol√≥gica excesiva"
        }
    },
    {
        "Nombre Cient√≠fico": "Galinsoga parviflora",
        "Nombre Com√∫n": "Guasca",
        "Familia": "Asteraceae",
        "Principios Activos": "Flavonoides, saponinas, aceites esenciales",
        "Usos Tradicionales": "Digestivo, antiinflamatorio, ingrediente en ajiaco",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n anticoagulantes",
        "Interacciones": {
            "Anticoagulantes": "üü° RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Cient√≠fico": "Justicia pectoralis",
        "Nombre Com√∫n": "Tilo",
        "Familia": "Acanthaceae",
        "Principios Activos": "Cumarinas, flavonoides, alcaloides",
        "Usos Tradicionales": "Ansiol√≠tico, sedante suave, antitusivo",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n GABA",
        "Interacciones": {
            "Benzodiazepinas": "üî¥ ALTO RIESGO: Depresi√≥n respiratoria",
            "Alcohol": "üü° RIESGO MODERADO: Potenciaci√≥n efectos sedantes"
        }
    },
    {
        "Nombre Cient√≠fico": "Barnadesia spinosa",
        "Nombre Com√∫n": "Chilca",
        "Familia": "Asteraceae",
        "Principios Activos": "Sesquiterpenos, flavonoides, alcaloides",
        "Usos Tradicionales": "Antiinflamatorio, diur√©tico, tratamiento renal",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n diur√©ticos",
        "Interacciones": {
            "Diur√©ticos": "üü° RIESGO MODERADO: Desequilibrio electrol√≠tico",
            "Litio": "üî¥ ALTO RIESGO: Toxicidad aumentada"
        }
    },
    {
        "Nombre Cient√≠fico": "Desmodium molliculum",
        "Nombre Com√∫n": "Pega Pega",
        "Familia": "Fabaceae",
        "Principios Activos": "Alcaloides, flavonoides, taninos",
        "Usos Tradicionales": "Antiulceroso, relajante muscular, antiespasm√≥dico",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n relajantes musculares",
        "Interacciones": {
            "Relajantes musculares": "üü° RIESGO MODERADO: Sedaci√≥n excesiva"
        }
    },
    {
        "Nombre Cient√≠fico": "Ocimum basilicum",
        "Nombre Com√∫n": "Albahaca Negra",
        "Familia": "Lamiaceae",
        "Principios Activos": "Eugenol, linalool, metil chavicol",
        "Usos Tradicionales": "Carminativo, antiespasm√≥dico, antis√©ptico",
        "Mecanismo Interacci√≥n": "Inhibici√≥n agregaci√≥n plaquetaria",
        "Interacciones": {
            "Anticoagulantes": "üü° RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Cient√≠fico": "Erythroxylum coca",
        "Nombre Com√∫n": "Coca",
        "Familia": "Erythroxylaceae",
        "Principios Activos": "Coca√≠na, ecgonina, hojas ricas en minerales",
        "Usos Tradicionales": "Estimulante, analg√©sico, supresor del apetito",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n sistema nervioso simp√°tico",
        "Interacciones": {
            "Anfetaminas": "üî¥ ALTO RIESGO: Crisis hipertensiva",
            "Antidepresivos IMAO": "üî¥ ALTO RIESGO: Crisis serotonin√©rgica"
        }
    },
    {
        "Nombre Cient√≠fico": "Brugmansia arborea",
        "Nombre Com√∫n": "Borrachero",
        "Familia": "Solanaceae",
        "Principios Activos": "Alcaloides trop√°nicos (escopolamina, atropina)",
        "Usos Tradicionales": "Sedante, analg√©sico, rituales tradicionales",
        "Mecanismo Interacci√≥n": "Antagonismo colin√©rgico",
        "Interacciones": {
            "Anticolin√©rgicos": "üî¥ ALTO RIESGO: S√≠ndrome anticolin√©rgico",
            "Antihistam√≠nicos": "üü° RIESGO MODERADO: Sedaci√≥n potenciada"
        }
    },
    {
        "Nombre Cient√≠fico": "Sambucus nigra",
        "Nombre Com√∫n": "Sa√∫co",
        "Familia": "Adoxaceae",
        "Principios Activos": "Flavonoides, antocianinas, √°cidos org√°nicos",
        "Usos Tradicionales": "Antigripal, inmunoestimulante, diur√©tico",
        "Mecanismo Interacci√≥n": "Diur√©tico suave",
        "Interacciones": {
            "Diur√©ticos": "üü° RIESGO MODERADO: Potenciaci√≥n efecto diur√©tico"
        }
    },
    {
        "Nombre Cient√≠fico": "Taraxacum officinale",
        "Nombre Com√∫n": "Diente de Le√≥n",
        "Familia": "Asteraceae",
        "Principios Activos": "Taraxacina, inulina, flavonoides",
        "Usos Tradicionales": "Diur√©tico, depurativo hep√°tico, digestivo",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n diur√©ticos",
        "Interacciones": {
            "Diur√©ticos": "üü° RIESGO MODERADO: Desequilibrio electrol√≠tico",
            "Litio": "üî¥ ALTO RIESGO: Toxicidad aumentada"
        }
    },
    {
        "Nombre Cient√≠fico": "Calendula officinalis",
        "Nombre Com√∫n": "Cal√©ndula",
        "Familia": "Asteraceae",
        "Principios Activos": "Carotenoides, flavonoides, aceites esenciales",
        "Usos Tradicionales": "Cicatrizante, antiinflamatorio, antis√©ptico",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n anticoagulantes",
        "Interacciones": {
            "Anticoagulantes": "üü° RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Cient√≠fico": "Guazuma ulmifolia",
        "Nombre Com√∫n": "Gu√°cimo",
        "Familia": "Malvaceae",
        "Principios Activos": "Taninos, flavonoides, proantocianidinas",
        "Usos Tradicionales": "Antidiarreico, antiinflamatorio, tratamiento respiratorio",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n antibi√≥ticos",
        "Interacciones": {
            "Antibi√≥ticos": "üü¢ BAJO RIESGO: Posible potenciaci√≥n"
        }
    },
    {
        "Nombre Cient√≠fico": "Lippia alba",
        "Nombre Com√∫n": "Pronto Alivio",
        "Familia": "Verbenaceae",
        "Principios Activos": "Citral, limoneno, carvona",
        "Usos Tradicionales": "Analg√©sico, antiespasm√≥dico, sedante suave",
        "Mecanismo Interacci√≥n": "Potenciaci√≥n sedantes",
        "Interacciones": {
            "Benzodiazepinas": "üü° RIESGO MODERADO: Sedaci√≥n excesiva",
            "Alcohol": "üî¥ ALTO RIESGO: Depresi√≥n SNC"
        }
    }
]

# üåø FUNCI√ìN PARA NORMALIZAR TEXTO
def normalizar_texto(texto):
    sustituciones = {
        '√°': 'a', '√©': 'e', '√≠': 'i', '√≥': 'o', '√∫': 'u',
        '√Å': 'A', '√â': 'E', '√ç': 'I', '√ì': 'O', '√ö': 'U',
        '√º': 'u', '√ú': 'U'
    }
    texto = texto.lower()
    for original, remplazo in sustituciones.items():
        texto = texto.replace(original, remplazo)
    return texto.strip()

# üåà SISTEMA DE ALERTAS VISUALES CON EMOJIS DE BOLITAS DE COLOR
def formatear_interaccion(info):
    if "üî¥ ALTO RIESGO" in info:
        return "üî¥ " + info.replace("üî¥ ALTO RIESGO:", "")
    elif "üü° RIESGO MODERADO" in info:
        return "üü° " + info.replace("üü° RIESGO MODERADO:", "")
    elif "üü¢ BAJO RIESGO" in info:
        return "üü¢ " + info.replace("üü¢ BAJO RIESGO:", "")
    else:
        return "‚úÖ No se detectaron interacciones peligrosas"

# üåø FUNCIONES DEL SISTEMA
def mostrar_bienvenida():
    print("\n" + "="*70)
    print("üåøüíä SISTEMA DE INTERACCIONES PLANTA-MEDICAMENTO (BotaniCheck) üíäüåø")
    print("="*70)
    print("\n'Preservando la salud a trav√©s del conocimiento ancestral y cient√≠fico'")
    print("\nüîç Descubre interacciones entre plantas medicinales colombianas y f√°rmacos")


    # Leyenda de riesgos
    print("\nCLASIFICACI√ìN DE RIESGO:")
    print("üî¥ ALTO RIESGO: Interacci√≥n peligrosa - Evitar combinaci√≥n")
    print("üü° RIESGO MODERADO: Precauci√≥n requerida - Consultar especialista")
    print("üü¢ BAJO RIESGO: Riesgo m√≠nimo - Monitoreo recomendado")
    print("‚úÖ SIN RIESGO: No se detectaron interacciones conocidas")

def consultar_interaccion():
    print("\n" + "="*60)
    print("üåøüíä CONSULTA DE INTERACCIONES")
    print("="*70)

    # Primero pedimos la planta
    entrada = input("\n‚û°Ô∏è Ingrese nombre de la planta: ").strip()
    if not entrada:
        return

    planta_encontrada = None
    entrada_normalizada = normalizar_texto(entrada)

    # Buscar planta por nombre
    for p in plantas_db:
        nombre_comun = normalizar_texto(p["Nombre Com√∫n"])
        nombre_cientifico = normalizar_texto(p["Nombre Cient√≠fico"])
        if entrada_normalizada in nombre_comun or entrada_normalizada in nombre_cientifico:
            planta_encontrada = p
            break

    # Si no se encuentra, mostrar lista de plantas numerada
    if not planta_encontrada:
        print("\n‚ö†Ô∏è Planta no encontrada. Estas son las plantas disponibles:")
        for i, planta in enumerate(plantas_db, 1):
            print(f" {i}. {planta['Nombre Com√∫n']}")

        try:
            num = int(input("\n‚û°Ô∏è Seleccione el n√∫mero de la planta: "))
            if 1 <= num <= len(plantas_db):
                planta_encontrada = plantas_db[num-1]
            else:
                print("\n‚ö†Ô∏è N√∫mero fuera de rango. Volviendo al men√∫ principal.")
                return
        except ValueError:
            print("\n‚ö†Ô∏è Entrada inv√°lida. Volviendo al men√∫ principal.")
            return

    # Ahora pedimos el medicamento
    medicamento = input("‚û°Ô∏è Ingrese nombre del medicamento: ").strip()
    if not medicamento:
        return

    medicamento_normalizado = normalizar_texto(medicamento)

    print("\n" + "="*60)
    print(f"üîç Resultados para: {planta_encontrada['Nombre Com√∫n']} + {medicamento}")
    print("="*60)

    # Buscar interacci√≥n
    interaccion_encontrada = None
    for med, info in planta_encontrada["Interacciones"].items():
        if medicamento_normalizado in normalizar_texto(med):
            interaccion_encontrada = (med, info)
            break

    # Mostrar resultados
    print(f"\nüåø {planta_encontrada['Nombre Com√∫n']} ({planta_encontrada['Nombre Cient√≠fico']})")
    print(f"üè• Familia: {planta_encontrada['Familia']}")
    print(f"üß™ Principios Activos: {planta_encontrada['Principios Activos']}")
    print(f"üí° Usos Tradicionales: {planta_encontrada['Usos Tradicionales']}")

    if interaccion_encontrada:
        med, info = interaccion_encontrada
        print(f"\n‚ö° INTERACCI√ìN DETECTADA CON {med.upper()}")
        print(formatear_interaccion(info))
        print(f"\n‚öôÔ∏è Mecanismo: {planta_encontrada['Mecanismo Interacci√≥n']}")
    else:
        print("\n" + formatear_interaccion("SIN RIESGO"))
        print("‚ÑπÔ∏è Esto no garantiza seguridad absoluta. Consulte a un especialista")

    print("\n" + "="*60)

def listar_plantas():
    print("\n" + "="*60)
    print("üåø LISTA COMPLETA DE PLANTAS MEDICINALES üåø")
    print("="*70)

    print("\nPlantas disponibles en el sistema:\n")
    for i, planta in enumerate(plantas_db, 1):
        print(f"{i}. {planta['Nombre Com√∫n']} ({planta['Nombre Cient√≠fico']})")
        print(f"   üè• Familia: {planta['Familia']}")
        print(f"   üí° Usos: {planta['Usos Tradicionales'][:70]}...")
        print()

    print("\nTotal plantas registradas:", len(plantas_db))
    print("="*70)

def buscar_planta():
    print("\n" + "="*70)
    print("üîç BUSCAR PLANTA POR NOMBRE")
    print("="*70)

    nombre = input("\n‚û°Ô∏è Ingrese nombre de la planta: ").strip()
    if not nombre:
        return

    nombre_normalizado = normalizar_texto(nombre)

    coincidencias = []
    for p in plantas_db:
        nombre_comun = normalizar_texto(p["Nombre Com√∫n"])
        nombre_cientifico = normalizar_texto(p["Nombre Cient√≠fico"])
        if nombre_normalizado in nombre_comun or nombre_normalizado in nombre_cientifico:
            coincidencias.append(p)

    print("\n" + "="*70)
    print(f"üîç RESULTADOS DE B√öSQUEDA: '{nombre}'")
    print("="*70)

    if not coincidencias:
        print("\n‚ö†Ô∏è No se encontraron plantas con ese nombre")
        return

    for planta in coincidencias:
        print(f"\nüåø {planta['Nombre Com√∫n']} ({planta['Nombre Cient√≠fico']})")
        print(f"üè• Familia: {planta['Familia']}")
        print(f"üß™ Principios Activos: {planta['Principios Activos']}")
        print(f"üí° Usos Tradicionales: {planta['Usos Tradicionales']}")
        print(f"‚öôÔ∏è Mecanismo de Interacci√≥n: {planta['Mecanismo Interacci√≥n']}")

        if planta["Interacciones"]:
            print("\n‚ö†Ô∏è INTERACCIONES REGISTRADAS:")
            for med, info in planta["Interacciones"].items():
                print(f"   ‚Ä¢ {med}: {formatear_interaccion(info)}")
        else:
            print("\n" + formatear_interaccion("SIN RIESGO"))

        print("\n" + "-"*70)

    print("\nFin de resultados")
    print("="*60)

def informacion_proyecto():
    print("\n" + "="*70)
    print("üìö INFORMACI√ìN DEL PROYECTO")
    print("="*60)

    print("\nSistema Confiable de Prevenci√≥n de Interacciones entre Plantas Medicinales y F√°rmacos")
    print("\nüåé Este proyecto integra el conocimiento ancestral de la medicina tradicional")
    print("colombiana con la farmacolog√≠a moderna para prevenir interacciones peligrosas")

    print("\nüéØ Objetivos:")
    print("‚Ä¢ Documentar 20 plantas medicinales colombianas")
    print("‚Ä¢ Identificar interacciones con f√°rmacos de uso com√∫n")
    print("‚Ä¢ Crear un sistema accesible con alertas visuales")
    print("‚Ä¢ Preservar conocimiento etnom√©dico")

    print("\nüî¨ Metodolog√≠a:")
    print("‚Ä¢ Revisi√≥n de literatura cient√≠fica")
    print("‚Ä¢ An√°lisis fitoqu√≠mico")
    print("‚Ä¢ Documentaci√≥n de casos cl√≠nicos")
    print("‚Ä¢ Validaci√≥n con expertos")

    print("\nüë• Equipo:")
    print("‚Ä¢ Investigadores en farmacolog√≠a")
    print("‚Ä¢ Etnobot√°nicos")
    print("‚Ä¢ Comunidades tradicionales")

    print("\nüìå Contacto:")
    print("alezavi@gmail.com, katherine.duquev@gamil.com, lauraceballosg@gmail.com")

    print("\n" + "="*70)

def menu_principal():
    while True:
        print("\n" + "="*70)
        print("üåø MEN√ö PRINCIPAL üåø")
        print("="*70)
        print("1. üåøüíä Consultar interacci√≥n planta-medicamento")
        print("2. üìú Ver lista completa de plantas")
        print("3. üîç Buscar informaci√≥n de una planta")
        print("4. üìö Informaci√≥n del proyecto")
        print("0. üö™ Salir del sistema")
        print("="*70)

        opcion = input("\n‚û°Ô∏è Seleccione una opci√≥n: ").strip()

        if opcion == "1":
            consultar_interaccion()
        elif opcion == "2":
            listar_plantas()
        elif opcion == "3":
            buscar_planta()
        elif opcion == "4":
            informacion_proyecto()
        elif opcion == "0":
            print("\n" + "="*70)
            print("üåø ¬°Gracias por usar nuestro sistema! üåø")
            print("Que el conocimiento ancestral y cient√≠fico gu√≠e tu camino hacia la salud")
            print("="*70)
            break
        else:
            print("\n‚ö†Ô∏è Opci√≥n no v√°lida. Por favor seleccione 1-4 o 0 para salir.")

if __name__ == "__main__":
    mostrar_bienvenida()
    menu_principal()
```
