```python
# -*- coding: utf-8 -*-
"""
Sistema Interactivo de Prevención de Interacciones entre Plantas Medicinales y Fármacos
"""

# 🧪 BASE DE DATOS CON LAS PLANTAS SOLICITADAS
plantas_db = [
    {
        "Nombre Científico": "Croton lechleri",
        "Nombre Común": "Sangre de Drago",
        "Familia": "Euphorbiaceae",
        "Principios Activos": "Taspina, dimetilcedrusina, proantocianidinas",
        "Usos Tradicionales": "Cicatrizante, antiinflamatorio, antiviral",
        "Mecanismo Interacción": "Inhibición del citocromo P450",
        "Interacciones": {
            "Warfarina": "🔴 ALTO RIESGO: Hemorragias (Evidencia: 24 casos)",
            "Aspirina": "🔴 ALTO RIESGO: Potenciación anticoagulante",
            "Heparina": "🟡 RIESGO MODERADO: Aumento tiempo de sangrado"
        }
    },
    {
        "Nombre Científico": "Banisteriopsis caapi",
        "Nombre Común": "Yagé",
        "Familia": "Malpighiaceae",
        "Principios Activos": "Harmina, harmalina, tetrahidroharmina",
        "Usos Tradicionales": "Medicina tradicional, tratamiento de depresión",
        "Mecanismo Interacción": "Inhibición de la monoaminooxidasa",
        "Interacciones": {
            "Antidepresivos ISRS": "🔴 ALTO RIESGO: Síndrome serotoninérgico",
            "Dextrometorfano": "🔴 ALTO RIESGO: Toxicidad serotoninérgica"
        }
    },
    {
        "Nombre Científico": "Ilex guayusa",
        "Nombre Común": "Guayusa",
        "Familia": "Aquifoliaceae",
        "Principios Activos": "Cafeína, teobromina, L-teanina",
        "Usos Tradicionales": "Estimulante, antioxidante, regulador glucémico",
        "Mecanismo Interacción": "Inhibición CYP1A2",
        "Interacciones": {
            "Clozapina": "🔴 ALTO RIESGO: Toxicidad aumentada",
            "Teofilina": "🟡 RIESGO MODERADO: Potenciación efectos"
        }
    },
    {
        "Nombre Científico": "Gliricidia sepium",
        "Nombre Común": "Matarratón",
        "Familia": "Fabaceae",
        "Principios Activos": "Cumarinas, flavonoides, alcaloides",
        "Usos Tradicionales": "Antiparasitario, insecticida, tratamiento de piel",
        "Mecanismo Interacción": "Inhibición CYP3A4",
        "Interacciones": {
            "Estatinas": "🔴 ALTO RIESGO: Rabdomiólisis",
            "Anticoagulantes": "🟡 RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Científico": "Maytenus laevis",
        "Nombre Común": "Chuchuguaza",
        "Familia": "Celastraceae",
        "Principios Activos": "Maytansina, triterpenos, polifenoles",
        "Usos Tradicionales": "Afrodisíaco, antiinflamatorio, tratamiento reumatismo",
        "Mecanismo Interacción": "Modulación sistema inmunológico",
        "Interacciones": {
            "Inmunosupresores": "🔴 ALTO RIESGO: Antagonismo terapéutico",
            "Corticosteroides": "🟡 RIESGO MODERADO: Reducción eficacia"
        }
    },
    {
        "Nombre Científico": "Petiveria alliacea",
        "Nombre Común": "Anamú",
        "Familia": "Phytolaccaceae",
        "Principios Activos": "Flavonoides, cumarinas, triterpenos",
        "Usos Tradicionales": "Antitumoral, analgésico, inmunoestimulante",
        "Mecanismo Interacción": "Inducción enzimática CYP450",
        "Interacciones": {
            "Anticoagulantes": "🔴 ALTO RIESGO: Sangrados",
            "Quimioterapéuticos": "🟡 RIESGO MODERADO: Potencial interacción"
        }
    },
    {
        "Nombre Científico": "Cissus verticillata",
        "Nombre Común": "Insulina Vegetal",
        "Familia": "Vitaceae",
        "Principios Activos": "Fitosteroles, flavonoides, ácidos grasos",
        "Usos Tradicionales": "Antidiabético, hipoglucemiante, antiinflamatorio",
        "Mecanismo Interacción": "Potenciación insulina",
        "Interacciones": {
            "Insulina": "🔴 ALTO RIESGO: Hipoglucemia severa",
            "Metformina": "🟡 RIESGO MODERADO: Potenciación efecto"
        }
    },
    {
        "Nombre Científico": "Uncaria tomentosa",
        "Nombre Común": "Uña de Gato",
        "Familia": "Rubiaceae",
        "Principios Activos": "Alcaloides oxindólicos, glicósidos, taninos",
        "Usos Tradicionales": "Inmunomodulador, antiinflamatorio, antiviral",
        "Mecanismo Interacción": "Estimulación sistema inmunológico",
        "Interacciones": {
            "Inmunosupresores": "🔴 ALTO RIESGO: Antagonismo terapéutico",
            "Vacunas vivas": "🟡 RIESGO MODERADO: Respuesta inmunológica excesiva"
        }
    },
    {
        "Nombre Científico": "Galinsoga parviflora",
        "Nombre Común": "Guasca",
        "Familia": "Asteraceae",
        "Principios Activos": "Flavonoides, saponinas, aceites esenciales",
        "Usos Tradicionales": "Digestivo, antiinflamatorio, ingrediente en ajiaco",
        "Mecanismo Interacción": "Potenciación anticoagulantes",
        "Interacciones": {
            "Anticoagulantes": "🟡 RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Científico": "Justicia pectoralis",
        "Nombre Común": "Tilo",
        "Familia": "Acanthaceae",
        "Principios Activos": "Cumarinas, flavonoides, alcaloides",
        "Usos Tradicionales": "Ansiolítico, sedante suave, antitusivo",
        "Mecanismo Interacción": "Potenciación GABA",
        "Interacciones": {
            "Benzodiazepinas": "🔴 ALTO RIESGO: Depresión respiratoria",
            "Alcohol": "🟡 RIESGO MODERADO: Potenciación efectos sedantes"
        }
    },
    {
        "Nombre Científico": "Barnadesia spinosa",
        "Nombre Común": "Chilca",
        "Familia": "Asteraceae",
        "Principios Activos": "Sesquiterpenos, flavonoides, alcaloides",
        "Usos Tradicionales": "Antiinflamatorio, diurético, tratamiento renal",
        "Mecanismo Interacción": "Potenciación diuréticos",
        "Interacciones": {
            "Diuréticos": "🟡 RIESGO MODERADO: Desequilibrio electrolítico",
            "Litio": "🔴 ALTO RIESGO: Toxicidad aumentada"
        }
    },
    {
        "Nombre Científico": "Desmodium molliculum",
        "Nombre Común": "Pega Pega",
        "Familia": "Fabaceae",
        "Principios Activos": "Alcaloides, flavonoides, taninos",
        "Usos Tradicionales": "Antiulceroso, relajante muscular, antiespasmódico",
        "Mecanismo Interacción": "Potenciación relajantes musculares",
        "Interacciones": {
            "Relajantes musculares": "🟡 RIESGO MODERADO: Sedación excesiva"
        }
    },
    {
        "Nombre Científico": "Ocimum basilicum",
        "Nombre Común": "Albahaca Negra",
        "Familia": "Lamiaceae",
        "Principios Activos": "Eugenol, linalool, metil chavicol",
        "Usos Tradicionales": "Carminativo, antiespasmódico, antiséptico",
        "Mecanismo Interacción": "Inhibición agregación plaquetaria",
        "Interacciones": {
            "Anticoagulantes": "🟡 RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Científico": "Erythroxylum coca",
        "Nombre Común": "Coca",
        "Familia": "Erythroxylaceae",
        "Principios Activos": "Cocaína, ecgonina, hojas ricas en minerales",
        "Usos Tradicionales": "Estimulante, analgésico, supresor del apetito",
        "Mecanismo Interacción": "Potenciación sistema nervioso simpático",
        "Interacciones": {
            "Anfetaminas": "🔴 ALTO RIESGO: Crisis hipertensiva",
            "Antidepresivos IMAO": "🔴 ALTO RIESGO: Crisis serotoninérgica"
        }
    },
    {
        "Nombre Científico": "Brugmansia arborea",
        "Nombre Común": "Borrachero",
        "Familia": "Solanaceae",
        "Principios Activos": "Alcaloides tropánicos (escopolamina, atropina)",
        "Usos Tradicionales": "Sedante, analgésico, rituales tradicionales",
        "Mecanismo Interacción": "Antagonismo colinérgico",
        "Interacciones": {
            "Anticolinérgicos": "🔴 ALTO RIESGO: Síndrome anticolinérgico",
            "Antihistamínicos": "🟡 RIESGO MODERADO: Sedación potenciada"
        }
    },
    {
        "Nombre Científico": "Sambucus nigra",
        "Nombre Común": "Saúco",
        "Familia": "Adoxaceae",
        "Principios Activos": "Flavonoides, antocianinas, ácidos orgánicos",
        "Usos Tradicionales": "Antigripal, inmunoestimulante, diurético",
        "Mecanismo Interacción": "Diurético suave",
        "Interacciones": {
            "Diuréticos": "🟡 RIESGO MODERADO: Potenciación efecto diurético"
        }
    },
    {
        "Nombre Científico": "Taraxacum officinale",
        "Nombre Común": "Diente de León",
        "Familia": "Asteraceae",
        "Principios Activos": "Taraxacina, inulina, flavonoides",
        "Usos Tradicionales": "Diurético, depurativo hepático, digestivo",
        "Mecanismo Interacción": "Potenciación diuréticos",
        "Interacciones": {
            "Diuréticos": "🟡 RIESGO MODERADO: Desequilibrio electrolítico",
            "Litio": "🔴 ALTO RIESGO: Toxicidad aumentada"
        }
    },
    {
        "Nombre Científico": "Calendula officinalis",
        "Nombre Común": "Caléndula",
        "Familia": "Asteraceae",
        "Principios Activos": "Carotenoides, flavonoides, aceites esenciales",
        "Usos Tradicionales": "Cicatrizante, antiinflamatorio, antiséptico",
        "Mecanismo Interacción": "Potenciación anticoagulantes",
        "Interacciones": {
            "Anticoagulantes": "🟡 RIESGO MODERADO: Sangrados"
        }
    },
    {
        "Nombre Científico": "Guazuma ulmifolia",
        "Nombre Común": "Guácimo",
        "Familia": "Malvaceae",
        "Principios Activos": "Taninos, flavonoides, proantocianidinas",
        "Usos Tradicionales": "Antidiarreico, antiinflamatorio, tratamiento respiratorio",
        "Mecanismo Interacción": "Potenciación antibióticos",
        "Interacciones": {
            "Antibióticos": "🟢 BAJO RIESGO: Posible potenciación"
        }
    },
    {
        "Nombre Científico": "Lippia alba",
        "Nombre Común": "Pronto Alivio",
        "Familia": "Verbenaceae",
        "Principios Activos": "Citral, limoneno, carvona",
        "Usos Tradicionales": "Analgésico, antiespasmódico, sedante suave",
        "Mecanismo Interacción": "Potenciación sedantes",
        "Interacciones": {
            "Benzodiazepinas": "🟡 RIESGO MODERADO: Sedación excesiva",
            "Alcohol": "🔴 ALTO RIESGO: Depresión SNC"
        }
    }
]

# 🌿 FUNCIÓN PARA NORMALIZAR TEXTO
def normalizar_texto(texto):
    sustituciones = {
        'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
        'Á': 'A', 'É': 'E', 'Í': 'I', 'Ó': 'O', 'Ú': 'U',
        'ü': 'u', 'Ü': 'U'
    }
    texto = texto.lower()
    for original, remplazo in sustituciones.items():
        texto = texto.replace(original, remplazo)
    return texto.strip()

# 🌈 SISTEMA DE ALERTAS VISUALES CON EMOJIS DE BOLITAS DE COLOR
def formatear_interaccion(info):
    if "🔴 ALTO RIESGO" in info:
        return "🔴 " + info.replace("🔴 ALTO RIESGO:", "")
    elif "🟡 RIESGO MODERADO" in info:
        return "🟡 " + info.replace("🟡 RIESGO MODERADO:", "")
    elif "🟢 BAJO RIESGO" in info:
        return "🟢 " + info.replace("🟢 BAJO RIESGO:", "")
    else:
        return "✅ No se detectaron interacciones peligrosas"

# 🌿 FUNCIONES DEL SISTEMA
def mostrar_bienvenida():
    print("\n" + "="*70)
    print("🌿💊 SISTEMA DE INTERACCIONES PLANTA-MEDICAMENTO (BotaniCheck) 💊🌿")
    print("="*70)
    print("\n'Preservando la salud a través del conocimiento ancestral y científico'")
    print("\n🔍 Descubre interacciones entre plantas medicinales colombianas y fármacos")


    # Leyenda de riesgos
    print("\nCLASIFICACIÓN DE RIESGO:")
    print("🔴 ALTO RIESGO: Interacción peligrosa - Evitar combinación")
    print("🟡 RIESGO MODERADO: Precaución requerida - Consultar especialista")
    print("🟢 BAJO RIESGO: Riesgo mínimo - Monitoreo recomendado")
    print("✅ SIN RIESGO: No se detectaron interacciones conocidas")

def consultar_interaccion():
    print("\n" + "="*60)
    print("🌿💊 CONSULTA DE INTERACCIONES")
    print("="*70)

    # Primero pedimos la planta
    entrada = input("\n➡️ Ingrese nombre de la planta: ").strip()
    if not entrada:
        return

    planta_encontrada = None
    entrada_normalizada = normalizar_texto(entrada)

    # Buscar planta por nombre
    for p in plantas_db:
        nombre_comun = normalizar_texto(p["Nombre Común"])
        nombre_cientifico = normalizar_texto(p["Nombre Científico"])
        if entrada_normalizada in nombre_comun or entrada_normalizada in nombre_cientifico:
            planta_encontrada = p
            break

    # Si no se encuentra, mostrar lista de plantas numerada
    if not planta_encontrada:
        print("\n⚠️ Planta no encontrada. Estas son las plantas disponibles:")
        for i, planta in enumerate(plantas_db, 1):
            print(f" {i}. {planta['Nombre Común']}")

        try:
            num = int(input("\n➡️ Seleccione el número de la planta: "))
            if 1 <= num <= len(plantas_db):
                planta_encontrada = plantas_db[num-1]
            else:
                print("\n⚠️ Número fuera de rango. Volviendo al menú principal.")
                return
        except ValueError:
            print("\n⚠️ Entrada inválida. Volviendo al menú principal.")
            return

    # Ahora pedimos el medicamento
    medicamento = input("➡️ Ingrese nombre del medicamento: ").strip()
    if not medicamento:
        return

    medicamento_normalizado = normalizar_texto(medicamento)

    print("\n" + "="*60)
    print(f"🔍 Resultados para: {planta_encontrada['Nombre Común']} + {medicamento}")
    print("="*60)

    # Buscar interacción
    interaccion_encontrada = None
    for med, info in planta_encontrada["Interacciones"].items():
        if medicamento_normalizado in normalizar_texto(med):
            interaccion_encontrada = (med, info)
            break

    # Mostrar resultados
    print(f"\n🌿 {planta_encontrada['Nombre Común']} ({planta_encontrada['Nombre Científico']})")
    print(f"🏥 Familia: {planta_encontrada['Familia']}")
    print(f"🧪 Principios Activos: {planta_encontrada['Principios Activos']}")
    print(f"💡 Usos Tradicionales: {planta_encontrada['Usos Tradicionales']}")

    if interaccion_encontrada:
        med, info = interaccion_encontrada
        print(f"\n⚡ INTERACCIÓN DETECTADA CON {med.upper()}")
        print(formatear_interaccion(info))
        print(f"\n⚙️ Mecanismo: {planta_encontrada['Mecanismo Interacción']}")
    else:
        print("\n" + formatear_interaccion("SIN RIESGO"))
        print("ℹ️ Esto no garantiza seguridad absoluta. Consulte a un especialista")

    print("\n" + "="*60)

def listar_plantas():
    print("\n" + "="*60)
    print("🌿 LISTA COMPLETA DE PLANTAS MEDICINALES 🌿")
    print("="*70)

    print("\nPlantas disponibles en el sistema:\n")
    for i, planta in enumerate(plantas_db, 1):
        print(f"{i}. {planta['Nombre Común']} ({planta['Nombre Científico']})")
        print(f"   🏥 Familia: {planta['Familia']}")
        print(f"   💡 Usos: {planta['Usos Tradicionales'][:70]}...")
        print()

    print("\nTotal plantas registradas:", len(plantas_db))
    print("="*70)

def buscar_planta():
    print("\n" + "="*70)
    print("🔍 BUSCAR PLANTA POR NOMBRE")
    print("="*70)

    nombre = input("\n➡️ Ingrese nombre de la planta: ").strip()
    if not nombre:
        return

    nombre_normalizado = normalizar_texto(nombre)

    coincidencias = []
    for p in plantas_db:
        nombre_comun = normalizar_texto(p["Nombre Común"])
        nombre_cientifico = normalizar_texto(p["Nombre Científico"])
        if nombre_normalizado in nombre_comun or nombre_normalizado in nombre_cientifico:
            coincidencias.append(p)

    print("\n" + "="*70)
    print(f"🔍 RESULTADOS DE BÚSQUEDA: '{nombre}'")
    print("="*70)

    if not coincidencias:
        print("\n⚠️ No se encontraron plantas con ese nombre")
        return

    for planta in coincidencias:
        print(f"\n🌿 {planta['Nombre Común']} ({planta['Nombre Científico']})")
        print(f"🏥 Familia: {planta['Familia']}")
        print(f"🧪 Principios Activos: {planta['Principios Activos']}")
        print(f"💡 Usos Tradicionales: {planta['Usos Tradicionales']}")
        print(f"⚙️ Mecanismo de Interacción: {planta['Mecanismo Interacción']}")

        if planta["Interacciones"]:
            print("\n⚠️ INTERACCIONES REGISTRADAS:")
            for med, info in planta["Interacciones"].items():
                print(f"   • {med}: {formatear_interaccion(info)}")
        else:
            print("\n" + formatear_interaccion("SIN RIESGO"))

        print("\n" + "-"*70)

    print("\nFin de resultados")
    print("="*60)

def informacion_proyecto():
    print("\n" + "="*70)
    print("📚 INFORMACIÓN DEL PROYECTO")
    print("="*60)

    print("\nSistema Confiable de Prevención de Interacciones entre Plantas Medicinales y Fármacos")
    print("\n🌎 Este proyecto integra el conocimiento ancestral de la medicina tradicional")
    print("colombiana con la farmacología moderna para prevenir interacciones peligrosas")

    print("\n🎯 Objetivos:")
    print("• Documentar 20 plantas medicinales colombianas")
    print("• Identificar interacciones con fármacos de uso común")
    print("• Crear un sistema accesible con alertas visuales")
    print("• Preservar conocimiento etnomédico")

    print("\n🔬 Metodología:")
    print("• Revisión de literatura científica")
    print("• Análisis fitoquímico")
    print("• Documentación de casos clínicos")
    print("• Validación con expertos")

    print("\n👥 Equipo:")
    print("• Investigadores en farmacología")
    print("• Etnobotánicos")
    print("• Comunidades tradicionales")

    print("\n📌 Contacto:")
    print("alezavi@gmail.com, katherine.duquev@gamil.com, lauraceballosg@gmail.com")

    print("\n" + "="*70)

def menu_principal():
    while True:
        print("\n" + "="*70)
        print("🌿 MENÚ PRINCIPAL 🌿")
        print("="*70)
        print("1. 🌿💊 Consultar interacción planta-medicamento")
        print("2. 📜 Ver lista completa de plantas")
        print("3. 🔍 Buscar información de una planta")
        print("4. 📚 Información del proyecto")
        print("0. 🚪 Salir del sistema")
        print("="*70)

        opcion = input("\n➡️ Seleccione una opción: ").strip()

        if opcion == "1":
            consultar_interaccion()
        elif opcion == "2":
            listar_plantas()
        elif opcion == "3":
            buscar_planta()
        elif opcion == "4":
            informacion_proyecto()
        elif opcion == "0":
            print("\n" + "="*70)
            print("🌿 ¡Gracias por usar nuestro sistema! 🌿")
            print("Que el conocimiento ancestral y científico guíe tu camino hacia la salud")
            print("="*70)
            break
        else:
            print("\n⚠️ Opción no válida. Por favor seleccione 1-4 o 0 para salir.")

if __name__ == "__main__":
    mostrar_bienvenida()
    menu_principal()
```
